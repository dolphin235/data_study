```python
import matplotlib.pyplot as plt

# Sample data for the line graph
x_values = years
y_values = list(df.iloc[93][years])

# Plot the line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Line Graph')

# Customize the plot
plt.title('Simple Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for the line graph
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

# Plot the line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Line Graph')

# Customize the plot
plt.title('Simple Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
# Sample data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 20, 25, 25]  # These are percentages, make sure they add up to 100

# Plot the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightpink'])

# Draw a circle at the center to make it look like a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Simple Pie Chart')

# Show the pie chart
plt.show()
```
